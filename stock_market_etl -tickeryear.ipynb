```python
import pandas as pd

```

## Extraction

```python
csv_file1 = "Resources/fundamentals.csv"
fundamentals_data_df = pd.read_csv(csv_file1)
fundamentals_data_df.head()
```

```python
csv_file2 = "Resources/prices.csv"
prices_data_df = pd.read_csv(csv_file2)
prices_data_df.head()
```

```python
csv_file3 = "Resources/securities.csv"
securities_data_df = pd.read_csv(csv_file3)
securities_data_df.head()
```

## Transformation

```python
securities_data_df.columns
```

```python
securities_data_df['Year'] = pd.DatetimeIndex(securities_data_df['Date first added']).year
securities_data_df['Year']
```

```python
dS = securities_data_df[["Ticker symbol",
                         "Security", 
                         "GICS Sector",
                         "GICS Sub Industry",
                         "Year"
                         ]]
dS
```

```python
prices_data_df.columns
```

```python
prices_data_df['year'] = pd.DatetimeIndex(prices_data_df['date']).year.astype('float')
prices_data_df['year']
```

```python
dP =prices_data_df [['symbol', 'open', 'close', 'low', 'high', 'volume','year']]
dP
```

```python
fundamentals_data_df.columns
```

```python
dF = fundamentals_data_df[["Ticker Symbol",                            
                           "Total Revenue", 
                           "Sales, General and Admin.", 
                           "Cost of Revenue",
                           "Research and Development",
                           "Earnings Before Interest and Tax",
                           "Operating Income",
                           "Net Income",   
                           "Earnings Per Share",
                           "Estimated Shares Outstanding",
                           "Total Assets",                            
                           "Total Liabilities",
                           "Total Equity",
                           "For Year" 
]]
dF
```

```python
merged_df = pd.merge(dF, dP, how='left', left_on=['Ticker Symbol','For Year'], right_on=['symbol','year'])
merged_df.head()
```

```python
new_merged_df = pd.merge(merged_df, dS, how='left', left_on=['Ticker Symbol','For Year'], right_on=['Ticker symbol','Year'])
new_merged_df.head()
```

```python
new_merged_df.count()
```

```python
new_df = new_merged_df.dropna(how='any')
new_df.count()
```

```python
new_df['Ticker Symbol'].nunique()
```

```python
drop_df = new_df.drop(columns=['Ticker symbol','symbol','For Year','year'])
drop_df
```

```python
drop_df.dtypes
```

```python
drop_df['Year'] = drop_df['Year'].astype('int') 
drop_df.dtypes
```

```python
drop_df['Total Revenue']= drop_df['Total Revenue']/1000000000
drop_df['Sales, General and Admin.']= drop_df['Sales, General and Admin.']/1000000000
drop_df['Cost of Revenue']= drop_df['Cost of Revenue']/1000000000
drop_df['Research and Development']= drop_df['Research and Development']/1000000000
drop_df['Earnings Before Interest and Tax']= drop_df['Earnings Before Interest and Tax']/1000000000
drop_df['Operating Income']= drop_df['Operating Income']/1000000000
drop_df['Net Income']= drop_df['Net Income']/1000000000
drop_df['Estimated Shares Outstanding']= drop_df['Estimated Shares Outstanding']/1000000000
drop_df['Total Assets']= drop_df['Total Assets']/1000000000
drop_df['Total Liabilities']= drop_df['Total Liabilities']/1000000000
drop_df['Total Equity']= drop_df['Total Equity']/1000000000
drop_df['volume']= drop_df['volume']/1000000
drop_df
```

```python
pd.set_option('display.float_format', '{:.3g}'.format)
drop_df
```

```python
drop_df.rename(columns={'Total Revenue':'Total Revenue $B',
                        'Sales, General and Admin.':'SG&A $B', 
                        'Cost of Revenue':'Cost of Revenue $B',
                        'Research and Development':'R&D $B',
                        'Earnings Before Interest and Tax':'EBIT $B', 
                        'Operating Income':'Operating Income $B', 
                        'Net Income':'Net Income $B', 
                        'Estimated Shares Outstanding':'Estimated Shares Outstanding $B', 
                        'Total Assets':'Total Assets $B', 
                        'Total Liabilities':'Total Liabilities $B', 
                        'Total Equity':'Total Equity $B',
                        'open':'Open',
                        'close':'Close',
                        'low':'Low',
                        'high':'High',
                        'volume':'Volume M',
                        
                        }, inplace=True)
    
drop_df
```

```python
drop_df.columns = drop_df.columns.str.replace(' ', '_')
drop_df
```

## Load

```python
# Use pandas to load csv converted DataFrame into database
drop_df.to_sql(name='nysedata', con=engine, if_exists='replace', index=False)
```
